## petition(국민청원)

```python
def get_petition(wd, num=777):
    base_url='https://www1.president.go.kr/petitions/'
    try:
        wd.get(base_url+str(num))
        element = WebDriverWait(wd, 2).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'petitionsView_title')))
        soup = bs(wd.page_source, 'lxml')

        progress = soup.find('div',{'class':'petitionsView_progress'}).get_text().strip()
        title = soup.find('h3',{'class':'petitionsView_title'}).get_text().strip()
        like = soup.find('h2',{'class':'petitionsView_count'}).get_text().strip()
        like = int(''.join(re.findall('\d+', str(like))))

        if progress == '답변완료':
            contents = soup.find_all('div',{'class':'View_write'})[1].get_text().strip()
            answer = soup.find_all('div',{'class':'pr_tk25'})[1].get_text().strip()[4:].strip()
        else:
            contents = soup.find('div',{'class':'View_write'}).get_text().strip()
            answer = ''

        info = soup.find('ul',{'class':'petitionsView_info_list'}).get_text().split('\n')
        category = info[1][4:]
        st_day = info[2][4:]
        ed_day = info[3][4:]
        proponent = info[4][3:].split()[0]
        return {'number':num, 'progress':progress, 'title':title,  'like':like,
                'contents':contents, 'answer':answer, 'category':category,
                'st_day':st_day, 'ed_day':ed_day, 'proponent':proponent}
    except Exception as e:
        print(int(num), e)
        return {'number':num, 'progress':'error', 'title':'',  'like':0,
                'contents':str(e), 'answer':'', 'category':'',
                'st_day':'', 'ed_day':'', 'proponent':''}
```

```python
def make_frame(crawling_def, options, num=777, cnt=10, init_num=200):
    rtn = []
    wd = webdriver.Chrome('chromedriver',options=options)
    for idx, i in enumerate(tqdm(range(num-cnt, num))):
        tmp = literal_eval(crawling_def+f'({wd},{i})')
        if idx == 0:
            rtn_c = list(tmp.keys())
        rtn.append(list(tmp.values()))
        # 오래 유지될경우 느려지는 경향이 있어 중간에 초기화
        if idx % init_num == 0:
            wd.quit()
            wd = webdriver.Chrome('chromedriver',options=options)
    return (rtn_c, rtn)


c, rtn = make_frame('get_petition', c_options, 50000, 10, 250)
%time wd = webdriver.Chrome('chromedriver',options=c_options)
%time rtn.append(list(get_petition(wd, 50000).values()))
%time rtn.append(list(get_petition(wd, 10).values()))
wd.quit()
pd.DataFrame(rtn, columns=c)
```
