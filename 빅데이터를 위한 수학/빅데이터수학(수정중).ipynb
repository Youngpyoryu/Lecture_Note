#1. 함수와 합성함수 (Function & Composition)

- 개념 요약: 함수는 입력값에 대응하는 출력값을 만드는 규칙

- 합성함수는 함수가 또 다른 함수의 입력으로 들어갈 때 만들어집니다.

```python
#1단계(기초) - 함수와 합성함수 시각화
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 2*x + 1
g = lambda x: x**2
x = np.linspace(-3, 3, 200)

plt.figure()
plt.plot(x, f(x), label='f(x)=2x+1')
plt.plot(x, g(x), label='g(x)=x^2')
plt.plot(x, g(f(x)), '--', label='g(f(x))')
plt.legend(); plt.grid(True); plt.title('Function Composition')
plt.show()
```

```python
#2단계(중급) - 입력에 따라 함수가 변할 때 / 함수의 계수 변화 -> 시각화
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 300)

# Try different slopes (a)
for a in [0.5, 1, 2]:
    f = lambda x: a*x + 1
    plt.plot(x, f(x), label=f'a={a}')

plt.legend(); plt.grid(True)
plt.title('Effect of slope a on f(x)=a*x+1')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.show()
```

```python
#3단계(응용) - 합성 순서 바꿔보기
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 2*x + 1
g = lambda x: x**2
x = np.linspace(-3, 3, 300)

plt.figure()
plt.plot(x, g(f(x)), '--', label='g(f(x))')
plt.plot(x, f(g(x)), '-.', label='f(g(x))')
plt.legend(); plt.grid(True)
plt.title('Compare g(f(x)) and f(g(x))')
plt.show()
```

```python
#4단계(확장)
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)
g = lambda x: x**2
x = np.linspace(-3, 3, 300)

plt.figure()
plt.plot(x, g(f(x)), '--', label='g(f(x))')
plt.plot(x, f(g(x)), '-.', label='f(g(x))')
plt.legend(); plt.grid(True)
plt.title('Compare g(f(x)) and f(g(x))')
plt.show()
```

#2. 로그와 지수함수 (Log & Exponential)
- 지수함수는 급격히 증가합니다.
- 로그함수는 서서히 증가합니다.
- 데이터 스케일이 너무 클 때 로그변환을 하면 분포를 보기 좋게 만들 수 있습니다.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 5, 200)
plt.figure()
plt.plot(x, np.exp(x), label='exp(x)')
plt.plot(x, np.log(x), label='log(x)')
plt.legend(); plt.grid(True); plt.title('Exponential vs Logarithm')
plt.show()
```

```python
#2단계(중급) - 축 스케일과 성질 이해
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 5, 200)

# (A) 같은 데이터, y축 로그 스케일
plt.figure()
plt.plot(x, np.exp(x), label='exp(x)')
plt.plot(x, np.log(x), label='log(x)')
plt.yscale('log')
plt.legend(); plt.grid(True); plt.title('Log-scale on Y (semilogy)')
plt.show()

# (B) x축과 y축을 각각 로그로 바꾸면 모양이 어떻게 달라지는지
plt.figure()
plt.plot(x, np.log(x), label='log(x)')
plt.xscale('log'); plt.yscale('linear')
plt.legend(); plt.grid(True); plt.title('Log-scale on X (semilogx)')
plt.show()
```

```python
#3단계(응용) - 데이터 분포, 변환 활용
##로그변환은 우측 치우친 분포를 다룰 때 유용.
## 거듭제곱 관계는 로그-로그 좌표에서 직선처럼 보임.

import numpy as np
import matplotlib.pyplot as plt

# (A) 로그정규분포: 원래 분포 vs 로그변환 후 분포
rng = np.random.default_rng(0)
data = rng.lognormal(mean=2.0, sigma=1.0, size=2000)

plt.figure()
plt.hist(data, bins=50)
plt.grid(True); plt.title('Skewed data (lognormal)')
plt.show()

plt.figure()
plt.hist(np.log(data), bins=50)
plt.grid(True); plt.title('After log transform: more symmetric')
plt.show()

# (B) 거듭제곱 관계 y = x^k 가 로그-로그에서 직선처럼 보이는지 확인
x = np.linspace(0.1, 10, 400)
y = x**2  # k=2

plt.figure()
plt.plot(x, y)
plt.grid(True); plt.title('y = x^2 (linear axes)')
plt.show()

plt.figure()
plt.plot(x, y)
plt.xscale('log'); plt.yscale('log')
plt.grid(True); plt.title('y = x^2 (log-log axes → straight line)')
plt.show()
```

#3.수열과 극한 (Sequence & Limit)

- 반복되는 값이 일정한 수로 가까워지면 ‘수렴’, 끝없이 커지면 ‘발산’이라고 합니다.

```python
#1단계(기초) - 수렴(수열이 일정한 값으로 가까워질때 "수렴(Converge)한다고 함.")
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 60)
a = 1 - 1/n
plt.plot(n, a, 'o-')
plt.axhline(1, linestyle='--')
plt.grid(True); plt.title('lim (1 - 1/n) = 1')
plt.show()
```
