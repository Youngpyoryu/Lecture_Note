##1. 함수와 합성함수 (Function & Composition)

- 개념 요약: 함수는 입력값에 대응하는 출력값을 만드는 규칙

- 합성함수는 함수가 또 다른 함수의 입력으로 들어갈 때 만들어집니다.

```python
#1단계(기초) - 함수와 합성함수 시각화
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 2*x + 1
g = lambda x: x**2
x = np.linspace(-3, 3, 200)

plt.figure()
plt.plot(x, f(x), label='f(x)=2x+1')
plt.plot(x, g(x), label='g(x)=x^2')
plt.plot(x, g(f(x)), '--', label='g(f(x))')
plt.legend(); plt.grid(True); plt.title('Function Composition')
plt.show()
```

함수 $f(x)=2x+1$과 $g(x)=x^2$를 각각 그리고, 두 함수를 합성한 $g(f(x))$의 그래프를 함께 시각화 해보자.

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 2*x + 1
g = lambda x: x**2
x = np.linspace(-3, 3, 200)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# 1번째: f(x)
axes[0].plot(x, f(x))
axes[0].set_title('f(x) = 2x + 1')
axes[0].grid(True)

# 2번째: g(x)
axes[1].plot(x, g(x))
axes[1].set_title('g(x) = x²')
axes[1].grid(True)

# 3번째: g(f(x))
axes[2].plot(x, g(f(x)), '--')
axes[2].set_title('g(f(x)) = (2x+1)²')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
# 3단계(응용) - 데이터 분포, 변환 활용
## 정규분포는 대칭이고, 로그변환하면 오히려 치우친다.
## Skewness(왜도): 정규분포에서 0에 가까움.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# (A) 정규분포: 원래 분포 vs 로그변환 후 분포
rng = np.random.default_rng(0)  # 난수 생성기
data = rng.normal(loc=0.0, scale=1.0, size=2000)  # 평균 0, 표준편차 1인 정규분포

# 원래(정규분포): 대칭적 형태
plt.figure()
plt.hist(data, bins=50, color='skyblue', edgecolor='black')
plt.grid(True)
plt.title('Original Normal Distribution (symmetric)')
plt.xlabel('Value'); plt.ylabel('Frequency')
plt.show()

# 왜도(Skewness) 계산
sk = skew(data)
print("Skewness (normal):", sk)

# 로그변환은 음수 불가능하므로 평행이동 후 적용
data_shifted = data - data.min() + 1e-6  # 전체를 양수로 이동
data_log = np.log(data_shifted)

# 로그변환 후: 오른쪽 꼬리가 긴 형태로 변형됨
plt.figure()
plt.hist(data_log, bins=50, color='salmon', edgecolor='black')
plt.grid(True)
plt.title('After log transform: now skewed (right tail)')
plt.xlabel('log(Value)'); plt.ylabel('Frequency')
plt.show()

# 왜도 비교
print("Skewness after log transform:", skew(data_log))
```

```python
#3단계(응용) - 데이터 분포, 변환 활용
##로그변환은 치우친 분포를 다룰 때 유용
## 거듭제곱 관계는 로그-로그좌표에서 직선처럼 보임.
## Skewness(왜도): 정규분포에 의해, 얼마나 치우쳐 있는지를 보는 것.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# (A) 로그정규분포: 원래 분포 vs 로그변환 후 분포
rng = np.random.default_rng(0)        # 난수(random number) 생성기 초기화
np.random.seed(42)                    # 시드 고정 → 동일한 난수 재현 가능
data = rng.lognormal(mean=2.0, sigma=1.0, size=2000)  # 로그정규분포 샘플 생성

# 원래(로그정규) 분포: 오른쪽 꼬리가 긴(skewed) 형태
plt.figure()
plt.hist(data, bins=50, color='skyblue', edgecolor='black')
plt.grid(True)
plt.title('Skewed data (Lognormal distribution)')
plt.xlabel('Value'); plt.ylabel('Frequency')
plt.show()

# 왜도(Skewness) 계산
sk = skew(data)
print("Skewness (lognormal):", sk)

# 로그변환 후: log(data) → 보다 대칭적인(정규에 가까운) 형태로 변환
plt.figure()
plt.hist(np.log(data), bins=50, color='salmon', edgecolor='black')
plt.grid(True)
plt.title('After log transform: more symmetric (≈ Normal)')
plt.xlabel('log(Value)'); plt.ylabel('Frequency')
plt.show()

# 왜도 비교
print("Skewness after log transform:", skew(np.log(data)))
```

##3.수열과 극한 (Sequence & Limit)

- 반복되는 값이 일정한 수로 가까워지면 ‘수렴’, 끝없이 커지면 ‘발산’이라고 합니다.

```python
#1단계(기초) - 수렴(수열이 일정한 값으로 가까워질때 "수렴(Converge)한다고 함.")
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 60)
a = 1 - 1/n
plt.plot(n, a, 'o-')
plt.axhline(1, linestyle='--')
plt.grid(True); plt.title('lim (1 - 1/n) = 1')
plt.show()
```

```python
# 2단계= - 발산과 교대수열
## 발산(diverge): 수열이 한 값으로 모이지 않음.
## 절대값은 수렴하지만 부호가 교대되는 교대수열.

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 20)
b = (-1)**n

plt.figure()
plt.stem(n, b, basefmt=" ")
plt.grid(True)
plt.title('Step 2 (Intermediate): Divergent sequence b_n = (-1)^n')
plt.xlabel('n'); plt.ylabel('b_n')
plt.show()
# 절댓값을 적용하면 |b_n| = 1 → 수렴하지 않음
# 반면 (-1)^n / n 은 절댓값은 감소하므로 0으로 수렴
c = (-1)**n / n

plt.figure()
plt.plot(n, c, 'o-', label='c_n = (-1)^n / n')
plt.axhline(0, linestyle='--', color='r', label='limit = 0')
plt.grid(True)
plt.title('Alternating sequence convergence: lim (-1)^n / n = 0')
plt.xlabel('n'); plt.ylabel('c_n')
plt.legend()
plt.show()
```

##4.미분과 변화율(Derivative & Rate of Change)

- 기울기는 함수가 얼마나 빨리 변하는지를 보여줌.

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2
x = np.linspace(-3, 3, 400)
a = 1.0

for dx in [1.0, 0.5, 0.1, 0.01]:
    slope = (f(a+dx) - f(a)) / dx
    plt.figure()
    plt.plot(x, f(x), label='f(x)=x^2')
    plt.plot([a, a+dx], [f(a), f(a+dx)], 'r--', label=f'slope≈{slope:.2f}')
    #≈ : windows (alt+247(키패드)) / Mac: Option+x
    plt.legend(); plt.grid(True)
    plt.title(f'Secant slope (dx={dx})')
    plt.xlabel('x'); plt.ylabel('f(x)')
    plt.show()
```

```python
#2단계 - 여러 점에서의 기울기
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2
x = np.linspace(-3, 3, 400)
points = [-2, -1, 0, 1, 2]

plt.figure()
plt.plot(x, f(x), label='f(x)=x^2')

for a in points:
    dx = 0.1
    slope = (f(a+dx) - f(a)) / dx
    tangent_x = np.linspace(a-0.8, a+0.8, 50)
    tangent_y = f(a) + slope * (tangent_x - a)
    plt.plot(tangent_x, tangent_y, '--', label=f'x={a}, slope≈{slope:.1f}')

plt.legend(); plt.grid(True)
plt.title('Tangents at different x positions')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.show()
```

##5.행렬의 곱
- 앞 행렬의 행(row)"와 뒤 행렬의 열(Coulmn)"을 각각 곱해 더한 것.


$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~C=A{\times}B$

단, A의 열수 = B의 행 수

```python
#1단계 - 원소별 곱(Element-wise Multiplication)
## * 연산자는 같은 위치의 원소끼리 곱하는 것.
## 수학적으로는 Hadamard product 또는 Element-wise product라고 함.

import numpy as np

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[10, 20],
              [30, 40]])

# Element-wise product
C = A * B
print("A =\n", A)
print("B =\n", B)
print("A * B (element-wise) =\n", C)
```

```python
#1단계- 행렬 곱의 기본 개념
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

C = A @ B   # same as np.dot(A,B)
print("A=\n", A)
print("B=\n", B)
print("A @ B =\n", C)
```

##6. 선형시스템 (Linear System) — 1단계 실습

연립방정식은 행렬로 표현할 수 있으며,  
파이썬에서는 `np.linalg.solve()`를 사용해 해를 구할 수 있습니다.

예를 들어 다음 식을 생각해 봅시다:


\begin{cases}
2x + y = 5 \\
x + 3y = 7
\end{cases}


이 식은 행렬 형태로 쓰면 이렇게 됩니다:


A =
\begin{bmatrix}
2 & 1 \\
1 & 3
\end{bmatrix},


b =
\begin{bmatrix}
5 \\
7
\end{bmatrix}

```python
import numpy as np

# Define matrix A and vector b
A = np.array([[2, 1],
              [1, 3]])
b = np.array([5, 7])

# Solve Ax = b
x = np.linalg.solve(A, b)

print("Matrix A:\n", A)
print("Vector b:\n", b)
print("Solution x:", x)
```

##7. 벡터 변환(Matrix Transformation)

```python
#1단계 - 벡터 변환
import numpy as np
import matplotlib.pyplot as plt

# Original vector
v = np.array([2, 1])

# Transformation matrix
A = np.array([[2, 0],
              [0, 1]])

# Transformed vector
v_new = A @ v

plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='Original v')
plt.quiver(0, 0, v_new[0], v_new[1], angles='xy', scale_units='xy', scale=1, color='r', label='Transformed A@v')
plt.axis('equal'); plt.grid(True)
plt.legend(); plt.title('Matrix Transform: Scaling along x-axis')
plt.show()
```

```python
#2단계 - 정사각형(square) 변환

import numpy as np
import matplotlib.pyplot as plt

# Define unit square (closed path)
square = np.array([[0, 1, 1, 0, 0],
                   [0, 0, 1, 1, 0]])

# Transformation matrix (rotation + shear)
A = np.array([[1, 0.5],
              [0, 1]])

# Apply transformation
square_new = A @ square

plt.figure()
plt.plot(square[0], square[1], 'b-', label='Original square')
plt.plot(square_new[0], square_new[1], 'r-', label='Transformed square')
plt.axis('equal'); plt.grid(True)
plt.legend(); plt.title('Linear Transformation of a Unit Square')
plt.xlabel('x'); plt.ylabel('y')
plt.show()
```

```python
#3단계 - 회전 행렬

import numpy as np
import matplotlib.pyplot as plt

# Original square
square = np.array([[0, 1, 1, 0, 0],
                   [0, 0, 1, 1, 0]])

# Rotation angle (30 degrees)
theta = np.pi / 6
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

# Apply rotation
square_rot = R @ square

plt.figure()
plt.plot(square[0], square[1], 'b-', label='Original square')
plt.plot(square_rot[0], square_rot[1], 'r-', label='Rotated by 30°')
plt.axis('equal'); plt.grid(True)
plt.legend(); plt.title('Rotation Transformation (30°)')
plt.xlabel('x'); plt.ylabel('y')
plt.show()
```

##8. 내적(Innter Product), 벡터의 놈(Norm), 벡터의 길이(length)


### 1. 내적 (Dot Product)

- 내적은 두 벡터가 얼마나 같은 방향을 향하는지 나타냅니다.  

$~~~~~~~~~\
\mathbf{a}\cdot\mathbf{b} = ||\mathbf{a}|| \, ||\mathbf{b}|| \cos\theta
\ $

- $ \theta\ $: 두 벡터 사이의 각도  
- 내적 > 0 → 두 벡터가 같은 방향  
- 내적 < 0 → 반대 방향  
- 내적 = 0 → 서로 직교 (수직)


```python
import numpy as np

a = np.array([2, 3])
b = np.array([4, 1])

dot = np.dot(a, b)
print("a · b =", dot)

# 각도 계산
theta = np.arccos(dot / (np.linalg.norm(a) * np.linalg.norm(b)))
print("Angle between a and b (degrees):", np.degrees(theta))
```

### 6. 벡터 투영

- 한 벡터를 다른 벡터가 이루는 직선 위로 수직으로 내려, 정사영(orthogonal projection)된 벡터를 구하는 것

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- vectors ----
a = np.array([2., 1.])
b = np.array([1., -1.])

# ---- calculations ----
dot = np.dot(a, b)
na, nb = np.linalg.norm(a), np.linalg.norm(b)
cos_th = dot / (na * nb)
theta_deg = np.degrees(np.arccos(cos_th))
u = b / nb
proj = np.dot(a, u) * u
resid = a - proj

# ---- plot ----
plt.figure(figsize=(7,7))
ax = plt.gca()
ax.set_aspect('equal', 'box')
plt.grid(True, linestyle=':', linewidth=0.6)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)

# base vectors
plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='#e74c3c',label='vector a',width=0.008)
plt.quiver(0,0,b[0],b[1],angles='xy',scale_units='xy',scale=1,color='#3498db',label='vector b',width=0.008)

# projection and residual
plt.quiver(0,0,proj[0],proj[1],angles='xy',scale_units='xy',scale=1,color='#2ecc71',label='projection of a on b',width=0.01)
plt.quiver(proj[0],proj[1],resid[0],resid[1],angles='xy',scale_units='xy',scale=1,color='#f1c40f',label='residual',width=0.01)

# connector line
plt.plot([proj[0],a[0]],[proj[1],a[1]],'--',color='gray')

# angle arc
ang = np.linspace(0,np.deg2rad(theta_deg),100)
r=1.2
plt.plot(r*np.cos(ang), r*np.sin(ang), color='purple', linestyle='--', linewidth=1)
plt.text(0.9*np.cos(np.deg2rad(theta_deg/2)), 0.9*np.sin(np.deg2rad(theta_deg/2)),
         f"{theta_deg:.1f}°", color='purple', fontsize=12, fontweight='bold')

# annotations
plt.text(a[0]*1.05, a[1]*1.05, 'a', fontsize=12, color='#e74c3c')
plt.text(b[0]*1.05, b[1]*1.05, 'b', fontsize=12, color='#3498db')
plt.text(proj[0]*1.05, proj[1]*1.05, 'proj', fontsize=12, color='#2ecc71')
plt.text((proj[0]+resid[0])*1.05, (proj[1]+resid[1])*1.05, 'a = proj + residual', fontsize=10)

plt.xlim(-3,3)
plt.ylim(-3,3)
plt.title("Vector Projection and Residual", fontsize=14)
plt.legend()
plt.show()
```

### 7. 최소자승법(Least Sqaure Method)

데이터 $(x_i,y_i)$들이 주어졌을 때, 모든 점을 정확히 지나는 직선을 찾기 어렵다면 오차 제곱합(Sum of Squared Errors)을 최소화하는 직선을 찾는다.

$~~~~~~~~~~~~~~~Ax ≈ y,~~~~~~ 즉,~~~min_x ||Ax-y||^2 $

$A$: 입력 데이터 행렬

$y$: 관측된 벡터

$x$: 추정할 계수 벡터

(해의 공식) $x_{Ls} = (A^TA)^{(-1)} A^Ty$

이는 정사영(Projection)관점에서, 벡터 $y$를  $A$의 열공간(Column space) 위로 투영(projection)한 결과

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- data ----------
np.random.seed(0)
n = 30
x = np.linspace(-3, 3, n)
m_true, c_true = 1.2, -0.7
y = m_true*x + c_true + np.random.normal(0, 0.8, n)

# ---------- design matrix A and LS solution ----------
A = np.c_[x, np.ones_like(x)]          # columns: [x, 1]
theta_ls, *_ = np.linalg.lstsq(A, y, rcond=None)  # [m_hat, c_hat]
m_hat, c_hat = theta_ls

# closed-form check
theta_cf = np.linalg.inv(A.T @ A) @ (A.T @ y)

# predictions, residuals, metrics
y_hat = A @ theta_ls
resid = y - y_hat
SSE = np.sum(resid**2)
SST = np.sum((y - y.mean())**2)
R2 = 1 - SSE/SST

# orthogonality check: A^T (y - y_hat) = 0
orth = A.T @ resid

print("[Least Squares]")
print("theta_ls  =", theta_ls)
print("theta_cf  =", theta_cf)
print("SSE =", SSE, "  R^2 =", R2)
print("A^T residual ≈ 0 ->", orth)

# ---------- plot ----------
plt.figure(figsize=(7,6))
plt.scatter(x, y, label="data")
xs = np.linspace(x.min(), x.max(), 200)
plt.plot(xs, m_hat*xs + c_hat, label="least-squares fit")
# residual lines
for xi, yi, yhi in zip(x, y, y_hat):
    plt.plot([xi, xi], [yi, yhi], linestyle='--', linewidth=0.8)

plt.title("Least Squares Fit: y ≈ m x + c")
plt.xlabel("x"); plt.ylabel("y")
plt.legend()
plt.text(0.02, 0.96, f"m={m_hat:.3f}, c={c_hat:.3f}\nSSE={SSE:.3f}, R²={R2:.3f}",
         transform=plt.gca().transAxes, va='top')
plt.show()
```

### 8 . 특이값 분해(Singular Value Decomposition,SVD)



#### (1) 기본 정의
모든 실수 행렬 $A \in \mathbb{R}^{m \times n}$ 은 다음과 같이 분해된다.

$$
A = U \Sigma V^\top
$$

- $U \in \mathbb{R}^{m \times m}$ : 직교행렬 (열벡터들이 서로 직교)
- $V \in \mathbb{R}^{n \times n}$ : 직교행렬
- $\Sigma \in \mathbb{R}^{m \times n}$ : 대각원소 $\sigma_1 \ge \sigma_2 \ge \cdots \ge \sigma_r > 0$

$r = \text{rank}(A)$ 이며, $\sigma_i$ 를 **특이값(singular value)** 이라 한다.

---

#### (2) 기하적 의미
SVD는 임의의 선형변환을 **회전 → 축척(스케일링) → 회전**으로 분해한 것이다.

$$
A x = U \Sigma V^\top x
$$

- $V^\top$ : 입력 좌표계를 회전  
- $\Sigma$ : 축 방향으로 특이값만큼 크기 조정  
- $U$ : 결과를 다시 회전  

---

#### (3) 랭크-$k$ 근사 (Eckart–Young 정리)

SVD 전개:
$$
A = \sum_{i=1}^{r} \sigma_i u_i v_i^\top
$$

상위 $k$개의 항만 남기면:
$$
A_k = \sum_{i=1}^{k} \sigma_i u_i v_i^\top
     = U_{:,1:k}\Sigma_{1:k,1:k}V_{:,1:k}^\top
$$

**Eckart–Young 정리**  
$$
A_k = \arg\min_{\text{rank}(B)=k} \|A - B\|_F
$$

즉, $A_k$는 Frobenius 노름 기준 **최적의 랭크-$k$ 근사**이다.

---

#### (4) 저랭크 근사의 의미
- 데이터의 **잡음 제거**, **중복 제거**
- **저장 공간 절약** (상위 $k$개만 저장)
- 영상의 경우 노이즈 억제와 유사한 효과

---

## 품질 평가 지표

### (1) 평균제곱오차 (MSE)
$$
\text{MSE} = \frac{1}{mn}\|A - A_k\|_F^2
$$

$k$가 커질수록 오차가 감소한다.

### (2) 피크 신호대 잡음비 (PSNR)
$$
\text{PSNR} = 10\log_{10}\!\left(\frac{\text{MAX}^2}{\text{MSE}}\right)
$$

- 데이터 범위가 $[0,1]$이면 $\text{MAX}=1$  
- $[0,255]$이면 $\text{MAX}=255$  
- 일반적으로  
  - 30 dB 이상 → 양호  
  - 40 dB 이상 → 매우 양호

---

## 주성분분석(PCA, Principal Component Analysis)

### (1) 개념
평균 제거된 데이터 $X_c \in \mathbb{R}^{n \times d}$ 에 대해

$$
X_c = U \Sigma V^\top
$$

- $V$의 열벡터 $v_i$ : 주성분 축  
- $\sigma_i^2$ : 각 주성분의 분산 크기

즉, PCA는 데이터의 **분산이 최대가 되는 방향**을 찾는 과정이다.

---

### (2) 설명분산비율 (Explained Variance Ratio)

$$
\text{Explained Ratio}_i = \frac{\sigma_i^2}{\sum_j \sigma_j^2}
$$

누적합으로 전체 분산 중 상위 $k$개가 차지하는 비율을 계산한다.

> 보통 누적 설명분산이 90~95% 이상이면 충분한 주성분 수로 본다.

---

### (3) 복원식 (Reconstruction)

상위 $k$개 주성분만으로 복원하면:

$$
X_k = (U_{:,1:k}\Sigma_{1:k,1:k})V_{:,1:k}^\top + \text{mean}
$$

$k$가 커질수록 원본과의 차이가 줄어든다.

---

##  SVD와 PCA의 관계

| 항목 | SVD | PCA |
|------|-----|-----|
| 입력 | 임의의 행렬 $A$ | 평균이 제거된 데이터 $X_c$ |
| 목적 | 저랭크 근사 (압축) | 분산 최대화 (축 찾기) |
| 축 벡터 | $V$의 열벡터 | $V$의 열벡터 (동일) |
| 특이값 | 데이터의 크기(√분산) | 주성분의 표준편차 |
| 응용 | 영상 압축, 노이즈 제거 | 차원 축소, 시각화 |

> PCA는 SVD의 특수한 형태이다.

---

## 실무적 기준 요약

| 구분 | 권장 기준 |
|------|------------|
| 영상 복원(SVD) | PSNR ≥ 30 dB : 양호 / ≥ 40 dB : 매우 양호 |
| 데이터 차원 축소(PCA) | 누적 설명분산 ≥ 90% → 적정 주성분 수 선택 |

---

> **요약:**  
> SVD는 행렬의 구조적 패턴을 가장 효율적으로 표현하는 방법이며,  
> PCA는 SVD를 이용해 데이터의 분산이 큰 축(주성분)을 찾아 차원을 줄이는 과정이다.


```python
import numpy as np
import matplotlib.pyplot as plt


def mse(a, b):
    # 평균제곱오차
    return np.mean((a - b)**2)

def psnr(a, b, max_val=1.0):
    # 피크 신호대 잡음비(PSNR)
    m = mse(a, b)
    if m == 0:
        return np.inf
    return 10*np.log10((max_val**2)/m)

# -----------------------------
# Part A. SVD 저랭크 근사 (이미지 행렬)
# -----------------------------
# 합성 "이미지" 생성: 평활 + 에지 + 잡음 → [0,1] 스케일
n, m = 256, 256
yy, xx = np.meshgrid(np.linspace(0,1,n), np.linspace(0,1,m), indexing='ij')
img = 0.4*np.sin(6*np.pi*xx)*np.cos(6*np.pi*yy) + 0.3*xx + 0.2*(yy>0.5).astype(float)
rng = np.random.default_rng(0)
img = img + 0.05*rng.standard_normal((n,m))
img = (img - img.min())/(img.max()-img.min())  # [0,1]

# SVD 분해
U, S, Vt = np.linalg.svd(img, full_matrices=False)  # img ≈ U @ diag(S) @ Vt

# 시험할 랭크들
ks = [5, 10, 20, 40, 80, 120]
mses, psnrs, recons = [], [], {}

for k in ks:
    # 상위 k개로 복원
    Uk, Sk, Vtk = U[:, :k], S[:k], Vt[:k, :]
    img_k = (Uk * Sk) @ Vtk  # U @ diag(S) @ Vt = (U * S) @ Vt (방식: 브로드캐스트)
    recons[k] = img_k
    mses.append(mse(img, img_k))
    psnrs.append(psnr(img, img_k, max_val=1.0))

# 원본과 몇 가지 랭크 복원 시각화(각각 단일 플롯로)
plt.figure(figsize=(4,4))
plt.imshow(img, cmap='gray'); plt.title("original"); plt.axis('off')
plt.show()

for k in ks[:3]:
    plt.figure(figsize=(4,4))
    idx = ks.index(k)
    plt.imshow(recons[k], cmap='gray')
    plt.title(f"rank {k}\nMSE={mses[idx]:.4f}\nPSNR={psnrs[idx]:.2f} dB")
    plt.axis('off')
    plt.show()

# MSE vs rank (단일 플롯)
plt.figure(figsize=(5,4))
plt.plot(ks, mses, 'o-')
plt.xlabel('rank k'); plt.ylabel('MSE'); plt.title('MSE vs rank')
plt.show()

# PSNR vs rank (단일 플롯)
plt.figure(figsize=(5,4))
plt.plot(ks, psnrs, 'o-')
plt.xlabel('rank k'); plt.ylabel('PSNR (dB)'); plt.title('PSNR vs rank')
plt.show()

print("[SVD low-rank image] ranks:", ks)
print("MSEs :", [f"{v:.5f}" for v in mses])
print("PSNRs:", [f"{v:.2f} dB" for v in psnrs])

# -----------------------------
# Part B. PCA via SVD (표 형식 데이터)
# -----------------------------
# 저차 잠재구조(3차원) + 잡음 → 50차원 관측 생성
rng = np.random.default_rng(1)
n_samples, n_features, true_k = 400, 50, 3
Z = rng.standard_normal((n_samples, true_k))          # 잠재요인
W = rng.standard_normal((true_k, n_features))         # 로딩
X = Z @ W + 0.2*rng.standard_normal((n_samples, n_features))  # 잡음 추가

# 평균 제거
mu = X.mean(axis=0)
Xc = X - mu

# SVD 기반 PCA
U2, S2, Vt2 = np.linalg.svd(Xc, full_matrices=False)
explained = (S2**2) / np.sum(S2**2)   # 설명분산비율
cumexp = np.cumsum(explained)

# k개 주성분으로 복원
k_pca = 3
Xc_k = (U2[:, :k_pca] * S2[:k_pca]) @ Vt2[:k_pca, :]
Xk = Xc_k + mu
recon_mse = mse(X, Xk)

print("\n[PCA via SVD]")
print("Top-6 explained variance ratio:", [f"{v:.3f}" for v in explained[:6]])
print(f"Cumulative (k={k_pca}): {cumexp[k_pca-1]:.3f}")
print(f"Reconstruction MSE (k={k_pca}): {recon_mse:.5f}")

# Scree plot (단일 플롯)
plt.figure(figsize=(6,4))
plt.plot(np.arange(1, 16), explained[:15], 'o-')
plt.xlabel('component'); plt.ylabel('explained var. ratio'); plt.title('Scree (first 15)')
plt.show()

# 누적 설명분산 (단일 플롯)
plt.figure(figsize=(6,4))
plt.plot(np.arange(1, 16), cumexp[:15], 'o-')
plt.xlabel('component'); plt.ylabel('cumulative ratio'); plt.title('Cumulative explained variance')
plt.show()
```

# 미적분학

## 1.극한과 도함수 정의 확인
$
f'(a) = \lim_{h\to0}\frac{f(a+h)-f(a)}{h}
$
도함수는 “순간 변화율”이며, 기하적으로는 접선의 기울기를 의미한다.



```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Numeric derivative via symmetric difference
def f(x): return np.where(x!=0, np.sin(x)/x, 1.0)
def fd(x, h=1e-4): return (f(x+h) - f(x-h)) / (2*h)

xs = np.linspace(-6, 6, 1000)
plt.figure(); plt.plot(xs, f(xs)); plt.title("f(x)=sin(x)/x with removable discontinuity"); plt.grid(True); plt.show()

# 2) Left/right limits at 0
for h in [1e-1, 1e-2, 1e-3, 1e-4]:
    print(h, f(0-h), f(0+h))

# 3) Tangent line at a=2
a = 2.0
m = fd(a)
b = f(a) - m*a
plt.figure()
plt.plot(xs, f(xs), label="f")
plt.plot(xs, m*xs + b, label=f"Tangent at x={a} (slope={m:.4f})")
plt.legend(); plt.grid(True); plt.show()
```

```python
# --- Stage 1: 극한과 도함수(정의) ---
import numpy as np
import matplotlib.pyplot as plt

# 0에서 제거가능한 불연속을 갖는 함수
def f(x):
    # 연속성 위해 f(0)=1로 정의
    return np.where(x != 0, np.sin(x)/x, 1.0)

# 1) a=0에서 좌/우 극한 확인
a = 0.0
hs = 10.0**(-np.arange(1, 8))  # 1e-1 ~ 1e-7
print("h        f(a-h)         f(a+h)")
for h in hs:
    print(f"{h:1.0e}   {f(a-h): .10f}   {f(a+h): .10f}")

# 2) 정의로 미분(전진 차분)
def deriv_def(a, h=1e-6):
    return (f(a + h) - f(a)) / h

a = 1.5
slope = deriv_def(a)

# 3) 함수와 접선 그리기
xs = np.linspace(-6, 6, 1200)
ys = f(xs)
tangent = slope*(xs - a) + f(a)

plt.figure()
plt.plot(xs, ys, label="f(x) = sin x / x")
plt.plot(xs, tangent, "--", label=f"Tangent at x={a}  slope≈{slope:.5f}")
plt.scatter([a], [f(a)], zorder=5)
plt.grid(True)
plt.legend()
plt.title("Limit and derivative by definition")
plt.show()
```

```python
import numpy as np
import sympy as sp

# --- MVT ---
def f(x): return x**3 - 2*x + 1
a, b = -1, 2
secant = (f(b)-f(a))/(b-a)
def fprime_num(x,h=1e-5): return (f(x+h)-f(x-h))/(2*h)

xs = np.linspace(a,b,6001)
vals = fprime_num(xs)
c = xs[np.argmin(np.abs(vals - secant))]
print(f"[MVT] secant={secant:.6f}, c≈{c:.6f}, f'(c)≈{fprime_num(c):.6f}")

# --- Rolle ---
def g(x): return (x-1)*(x+2)*(x-3)
ra, rb = -2, 1
xs = np.linspace(ra,rb,6001)
vals = fprime_num(xs)
print("[Rolle] c points:", xs[np.where(np.abs(vals)<1e-3)][:5])

# --- Chain Rule ---
x = sp.Symbol('x')
u = x**3 - 2*x
y = sp.exp(u)
lhs = sp.diff(y, x)
rhs = sp.diff(sp.exp(sp.Symbol('u')), sp.Symbol('u')).subs({'u':u}) * sp.diff(u, x)
print("[Chain rule] symbolic equality:", sp.simplify(lhs - rhs) == 0)
```

### 1. 산술평균, 기하평균, 조화평균



####산술평균 (Arithmetic Mean)

  여러 수의 합을 그 개수로 나눈 값  
  
  $
  A = \frac{1}{n}\sum_{i=1}^{n} x_i
  $

  - 가장 일반적인 평균
  - 극단값(Outlier)에 민감
  - 예: 시험 점수, 수익률의 단순 평균

---

#### 기하평균 (Geometric Mean)

여러 수의 곱의 \(n\)제곱근  
  $
  G = \left(\prod_{i=1}^{n} x_i\right)^{1/n}
  $


  - 모든 값이 양수일 때만 정의
  - 비율·성장률 계산에 적합
  - 로그 평균과 동일:  
    \[
    \ln G = \frac{1}{n}\sum_{i=1}^{n}\ln x_i
    \]
  - 예: 복리, 연평균 성장률 (CAGR)

---

#### 조화평균 (Harmonic Mean)
  역수의 평균의 역수  

  $
  H = \frac{n}{\sum_{i=1}^{n} (1/x_i)}
  $

  - 작은 값에 민감
  - 속도·비율 평균에 적합
  - 예: 평균 속도, 단가 계산

---

#### 관계식

$
H \le G \le A
$
- 항상 이 부등식이 성립  
- 모든 값이 같을 때 \(H = G = A\)

---

```python
import numpy as np
import matplotlib.pyplot as plt


data = np.array([2, 8, 12, 20])
n = len(data)


A = np.mean(data)
G = np.prod(data)**(1/n)
H = n / np.sum(1/data)

print("[Means]")
print(f"Data: {data}")
print(f"Arithmetic Mean = {A:.3f}")
print(f"Geometric Mean  = {G:.3f}")
print(f"Harmonic Mean   = {H:.3f}")


plt.figure(figsize=(8,3))
plt.bar(range(1, n+1), data, color='skyblue', label='data')
plt.axhline(A, color='r', linestyle='--', label=f'Arithmetic Mean = {A:.2f}')
plt.axhline(G, color='g', linestyle='--', label=f'Geometric Mean = {G:.2f}')
plt.axhline(H, color='orange', linestyle='--', label=f'Harmonic Mean = {H:.2f}')
plt.title("Comparison of Arithmetic / Geometric / Harmonic Means")
plt.legend()
plt.show()

# 극단값(outlier)영향 비교
outlier_data = np.append(data, 100)
A2 = np.mean(outlier_data)
G2 = np.prod(outlier_data)**(1/len(outlier_data))
H2 = len(outlier_data) / np.sum(1/outlier_data)

print("\n[With Outlier 100]")
print(f"Arithmetic Mean = {A2:.3f}")
print(f"Geometric Mean  = {G2:.3f}")
print(f"Harmonic Mean   = {H2:.3f}")

plt.figure(figsize=(8,3))
plt.bar(range(1, len(outlier_data)+1), outlier_data, color='lightgray', label='data with outlier')
plt.axhline(A2, color='r', linestyle='--', label=f'Arithmetic = {A2:.2f}')
plt.axhline(G2, color='g', linestyle='--', label=f'Geometric = {G2:.2f}')
plt.axhline(H2, color='orange', linestyle='--', label=f'Harmonic = {H2:.2f}')
plt.title("Effect of Outlier on Different Means")
plt.legend()
plt.show()
```

### 2. t-검정, 순열검정, F검정, 카이제곱검정

---

#### t-검정 (평균 비교)
- 두 집단의 **평균이 같은지** 비교.
- 데이터가 **정규분포**라고 가정.

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
  t = \frac{\bar{x}-\bar{y}}{\sqrt{\frac{s_x^2}{n_x}+\frac{s_y^2}{n_y}}}
$

- p값 < 0.05 → 평균이 다르다고 판단.

---

#### 순열검정 (평균 비교, 가정 없음)
- 두 집단 데이터를 **섞어서** 다시 나누고 평균 차이를 반복 계산.
- 이 과정을 여러 번(예: 5000번) 해서 “우연히 이런 차이가 날 확률”을 구함.
- **정규분포 가정이 필요 없음.**

---

#### F-검정 (분산 비교)
- 두 집단의 **데이터 흩어짐 정도(분산)** 이 같은지 비교.
- 정규분포 가정 필요.

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
  F = \frac{s_x^2}{s_y^2}
$

- p값 < 0.05 → 두 집단의 분산이 다름.

---

#### 카이제곱 검정 (한 집단 분산)
- 한 집단의 **분산이 특정 값(예: 1)** 과 같은지 확인.
- 정규분포 가정 필요.


  
$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\chi^2 = \frac{(n-1)s^2}{\sigma_0^2}$

- p값 < 0.05 → 분산이 다름.

```python
import numpy as np
from scipy.stats import t, f, chi2

rng = np.random.default_rng(123)

# 데이터
a = rng.normal(0, 1, 50)
b = rng.normal(0.4, 1, 60)

nx, ny = len(a), len(b)
mx, my = a.mean(), b.mean()
vx, vy = a.var(ddof=1), b.var(ddof=1)

# 1) Welch 두 표본 t-검정 (양측)
se2 = vx/nx + vy/ny
t_stat = (mx - my) / np.sqrt(se2)
df = se2**2 / ((vx/nx)**2/(nx-1) + (vy/ny)**2/(ny-1))
p_t = 2 * (1 - t.cdf(abs(t_stat), df))

print(f"[Welch t-test] t={t_stat:.4f}, df={df:.2f}, p={p_t:.4g}")

# 2) 순열검정 (양측)
obs = mx - my
comb = np.r_[a, b]
B = 5000
perm_stats = []
for _ in range(B):
    perm = rng.permutation(comb)
    perm_stats.append(perm[:nx].mean() - perm[nx:].mean())
perm_stats = np.array(perm_stats)
p_perm = (np.abs(perm_stats) >= abs(obs)).mean()

print(f"[Permutation] obs={obs:.4f}, p={p_perm:.4g}")

# 3) F검정: 등분산 (양측)
F = vx / vy
df1, df2 = nx-1, ny-1
# 양측 p-value: 두 꼬리 중 작은 쪽의 2배
p_F = 2 * min(f.cdf(F, df1, df2), 1 - f.cdf(F, df1, df2))

print(f"[F-test var equality] F={F:.4f}, df=({df1},{df2}), p={p_F:.4g}")

# 4) 카이제곱 검정: 단일 집단 분산 = sigma0^2 ?
sigma0_sq = 1.0  # 가설 분산(필요에 맞게 설정)
# 집단 a
chi_a = (nx-1)*vx / sigma0_sq
p_chi_a = 2 * min(chi2.cdf(chi_a, nx-1), 1 - chi2.cdf(chi_a, nx-1))
# 집단 b
chi_b = (ny-1)*vy / sigma0_sq
p_chi_b = 2 * min(chi2.cdf(chi_b, ny-1), 1 - chi2.cdf(chi_b, ny-1))

print(f"[Chi-square var=a] chi2={chi_a:.4f}, df={nx-1}, p={p_chi_a:.4g}")
print(f"[Chi-square var=b] chi2={chi_b:.4f}, df={ny-1}, p={p_chi_b:.4g}")
```

### 중심극한정리 (Central Limit Theorem, CLT)



- **많은 표본의 평균은 정규분포에 가까워진다**는 정리
- 모집단의 분포가 어떤 모양이든(비대칭, 한쪽으로 치우침 등),
  표본 크기 \(n\)이 충분히 크면 표본평균의 분포는 거의 정규분포가 됨.


모집단의 평균이 $\mu$, 분산이 $\sigma^2$일 때  
표본크기 $n$인 표본평균 $\bar{X}$는 $
\bar{X} \approx \mathcal{N}\left(\mu,\ \frac{\sigma^2}{n}\right)$


- 데이터의 분포가 비정규라도, **표본 평균의 분포는 정규로 수렴**.
- 그래서 t-검정이나 신뢰구간 계산처럼 “정규분포 가정”이 필요한 통계기법을 많은 표본에서 안심하고 쓸수 있음.

---

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)

def sample_fn(dist, size, **kwargs):
    """분포 이름과 매개변수로 표본 생성"""
    if dist == "uniform":      # U(a,b)
        a = kwargs.get("a", 0.0); b = kwargs.get("b", 1.0)
        return rng.uniform(a, b, size=size)
    if dist == "exponential":  # Exp(rate=alpha)  (평균=1/alph)
        alpha = kwargs.get("alpha", 1.0)
        return rng.exponential(1/alpha, size=size)
    if dist == "bernoulli":    # Ber(p)
        p = kwargs.get("p", 0.3)
        return rng.binomial(1, p, size=size).astype(float)
    raise ValueError("지원하지 않는 분포")

def clt_demo(dist="exponential", n=30, trials=5000, **kwargs):
    """
    dist: 'uniform' | 'exponential' | 'bernoulli'
    n: 표본 크기
    trials: 표본평균을 뽑는 반복 횟수
    kwargs: 각 분포의 매개변수
    """
    # 1) 모평균과 모분산(알려진 값 또는 모의추정)
    if dist == "uniform":
        a = kwargs.get("a", 0.0); b = kwargs.get("b", 1.0)
        mu, var = (a+b)/2, (b-a)**2/12
    elif dist == "exponential":
        lam = kwargs.get("lam", 1.0)
        mu, var = 1/lam, 1/(lam**2)
    elif dist == "bernoulli":
        p = kwargs.get("p", 0.3)
        mu, var = p, p*(1-p)
    else:
        raise ValueError("지원하지 않는 분포")

    # 2) 표본평균 생성
    means = np.empty(trials)
    for t in range(trials):
        x = sample_fn(dist, size=n, **kwargs)
        means[t] = x.mean()

    # 3) 정규근사 pdf
    mu_n = mu
    sigma_n = np.sqrt(var/n)
    xs = np.linspace(means.min(), means.max(), 400)
    normal_pdf = (1/np.sqrt(2*np.pi*sigma_n**2)) * np.exp(-(xs-mu_n)**2/(2*sigma_n**2))

    # 4) 시각화
    plt.figure()
    plt.hist(means, bins=40, density=True, alpha=0.6)
    plt.plot(xs, normal_pdf, linewidth=2)
    plt.title(f"CLT demo: dist={dist}, n={n}, trials={trials}")
    plt.xlabel("sample mean")
    plt.ylabel("density")
    plt.show()

    # 5) 수치 비교 출력
    print(f"sample mean ≈ {means.mean():.4f} (theory {mu_n:.4f})")
    print(f"sample std  ≈ {means.std(ddof=1):.4f} (theory {sigma_n:.4f})")

# 사용 예시
# 1) 지수분포(비대칭) → n 증가 시 정규로 수렴
clt_demo(dist="exponential", lam=0.5, n=5, trials=5000)
# clt_demo(dist="exponential", lam=0.5, n=30, trials=5000)

# 2) 균등분포
# clt_demo(dist="uniform", a=-2, b=5, n=20, trials=5000)

# 3) 베르누이
# clt_demo(dist="bernoulli", p=0.2, n=50, trials=5000)
```

### 몬테카를로(Monte Carlo)

- 난수를 이용해 확률적으로 어떤 문제의 근사해를 구하는 수치적 기법
- 수학적으로 해석적 계산이 어렵거나 복잡한 문제를 확률과 통계의 개념으로 단순화해 근사값을 구함.

**무작위 표본(Random Sampling)**을 통해 기댓값을 추정한다.

예를 들어, 함수 $f(x)$의 적분
$
I = \int_a^b f(x) \, dx
$
을 근사하려면 구간 $[a,b]$에서 난수 $ x_i $를 $ N $개 뽑아,
$
I \approx \frac{b - a}{N} \sum_{i=1}^{N} f(x_i)
$
로 계산한다.  
$ N $이 커질수록 결과는 실제값에 수렴한다.  
(이론적 근거: **큰수의 법칙**)


```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

rng = np.random.default_rng(42)

# --- Settings ---
# Significantly increase the total number of points for better convergence
TOTAL_POINTS = 500_000
batch = 5000
num_batches = TOTAL_POINTS // batch
xlim = ylim = (0, 1)

# --- Simulation ---
N_total = 0
inside_total = 0
estimates = []
N_history = []

for _ in range(num_batches):
    x = rng.random(batch)
    y = rng.random(batch)
    m = (x*x + y*y) <= 1.0

    N_total += batch
    inside_total += m.sum()

    estimates.append(inside_total / N_total)
    N_history.append(N_total)

# Store the last batch of points for the scatter plot
# or sample from all generated points if memory allows.
# For simplicity, we'll run a new, smaller batch for visualization only.
VIS_POINTS = 10000
x_vis = rng.random(VIS_POINTS)
y_vis = rng.random(VIS_POINTS)
m_vis = (x_vis*x_vis + y_vis*y_vis) <= 1.0

# --- Visualization ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle(f"Monte Carlo Estimation of $\\pi/4$ (Total $N={N_total:,}$)", fontsize=16)

# Subplot 1: Scatter Plot
ax1.set_title(f"Final Scatter Plot: Area $\\approx {inside_total / N_total:.6f}$")
ax1.set_xlim(*xlim)
ax1.set_ylim(*ylim)
ax1.set_aspect("equal")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

# Quarter circle reference
theta = np.linspace(0, np.pi/2, 400)
ax1.plot(np.cos(theta), np.sin(theta), "r--", lw=1.5, label="$x^2+y^2=1$")

# Scatter points (using a smaller, representative sample)
ax1.scatter(x_vis[m_vis], y_vis[m_vis], s=8, c="dodgerblue", label="Inside")
ax1.scatter(x_vis[~m_vis], y_vis[~m_vis], s=8, c="lightgray", label="Outside")
ax1.legend(loc="lower left")

# Subplot 2: Convergence History
true_val = np.pi / 4
ax2.plot(N_history, estimates, c='teal', label='MC Estimate')
ax2.axhline(true_val, color='r', linestyle='--', label='True Value ($\pi/4$)')
ax2.set_title("Convergence of the Area Estimate")
ax2.set_xlabel("Number of Samples ($N$)")
ax2.set_ylabel("Estimated Area")
ax2.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
ax2.legend()
ax2.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Visual convergence of quarter-circle area (π/4) without scatter — fast and clear
import numpy as np, matplotlib.pyplot as plt, time
from IPython.display import display

rng = np.random.default_rng(0)

# ---- settings ----
total_samples = 5_000_000   # 총 표본 수
batch = 100_000             # 한 번에 추가할 표본 수
steps = total_samples // batch
true_area = np.pi / 4       # 정답: 1/4 원의 넓이

# ---- state ----
N = 0
inside = 0
Ns, Est, Err = [], [], []

# ---- figure ----
plt.rcParams["figure.figsize"] = (9, 4.2)
fig, (ax_bar, ax_err) = plt.subplots(1, 2, gridspec_kw={"width_ratios":[1,1.2]})

# left: bar for estimated area vs true
bar = ax_bar.bar([0], [0.0], width=0.6, color="tab:blue")
ax_bar.axhline(true_area, color="k", ls="--", lw=1.5, label="true π/4")
ax_bar.set_ylim(0, 1.0)
ax_bar.set_xticks([])
ax_bar.set_ylabel("area in [0,1]²")
ax_bar.set_title("Estimated quarter-circle area")
ax_bar.legend(loc="upper right")

# right: absolute error vs N (log-x)
(line_err,) = ax_err.plot([], [], lw=2)
ax_err.set_xscale("log")
ax_err.set_xlim(1, total_samples)
ax_err.set_ylim(1e-6, 1e-1)
ax_err.set_xlabel("N (log scale)")
ax_err.set_ylabel("|estimate − π/4|")
ax_err.grid(True, which="both", ls=":")

disp = display(fig, display_id=True)
plt.close(fig)

t0 = time.perf_counter()
for step in range(steps):
    x = rng.random(batch); y = rng.random(batch)
    inside += np.count_nonzero(x*x + y*y <= 1.0)
    N += batch

    est = inside / N                   # 추정된 면적(π/4)
    err = abs(est - true_area)

    # 기록
    Ns.append(N); Est.append(est); Err.append(err)

    # 화면 갱신(가벼운 오버헤드로 자주 업데이트)
    if step % 2 == 0 or step == steps - 1:
        bar[0].set_height(est)
        line_err.set_data(Ns, Err)
        ax_err.set_ylim(max(min(Err)*0.5,1e-7), max(max(Err)*1.5,1e-5))  # 자동 스케일
        fig.suptitle(f"N={N:,}   area≈{est:.6f}   true={true_area:.6f}   |err|={err:.2e}   π≈{4*est:.6f}",
                     fontsize=12)
        disp.update(fig)

elapsed = time.perf_counter() - t0
print(f"Done: N={N:,}, area≈{Est[-1]:.6f}, true={true_area:.6f}, |err|={Err[-1]:.3e}, π≈{4*Est[-1]:.6f}, time={elapsed:.2f}s")
```

### 정규분포의 MLE, 신뢰구간, 부트스트랩

#### 1. 최대우도추정 (Maximum Likelihood Estimation, MLE)

- 정규분포  $ x_i \sim \mathcal{N}(\mu, \sigma^2) $ 의 확률밀도함수는  $ f(x_i|\mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}}
\exp\!\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)
$

- 로그우도함수(Log-likelihood): $
\ell(\mu, \sigma) = -\frac{n}{2}\log(2\pi) - n\log\sigma -
\frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2
$


이를 $\mu, \sigma$에 대해 미분해 0으로 두면,
$
\hat{\mu}_{MLE} = \bar{x}, \quad
\hat{\sigma}_{MLE} = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2}
$

즉, MLE는 **표본평균과 모분산 기준의 표본표준편차(ddof=0)** 임.

★ ddof, delta degrees of freedom
일반적으로 모집단의 표준 편차나 분산을 추정할 때 사용되는 추정량은 편향되지 않은(unbiased) 추정량입니다. 즉, 통계적 추정이 모집단의 실제 값에 가까워지도록 조정하는 것

---

####2. 모평균의 신뢰구간 (Confidence Interval)

표본평균의 표준오차(Standard Error)는$~~ SE = \frac{s}{\sqrt{n}}$

95% 신뢰구간은 (정규근사 가정하에) $~~~(\bar{x} - 1.96\times SE,\; \bar{x} + 1.96\times SE)$

---

####3. 부트스트랩 (Bootstrap)

- 데이터 분포를 가정하지 않고, 관측된 표본에서 **복원추출(resampling with replacement)** 하여  
  여러 개의 표본을 만든 뒤 각 표본의 통계량(예: 평균)을 계산.

- 목적: 추정량의 불확실성(분산, 신뢰구간)을 데이터 기반으로 추정.

- 절차:
  1. 원본 데이터에서 크기 $n$의 표본을 복원추출.
  2. 각 부트스트랩 표본의 평균 계산.
  3. 이 과정을 $B$번 반복 후, 평균들의 분포를 이용해 신뢰구간 추정.


```python
import numpy as np

# 1) 표본 생성 (정규분포 N(10, 2^2))
x = np.random.normal(10, 2, size=200)

# 2) MLE 추정
mu_mle = x.mean()              # MLE for mean
sigma_mle = x.std(ddof=0)      # MLE for std (모분산 기준)

# 3) 모평균의 95% 신뢰구간
ci_mu = (
    mu_mle - 1.96 * x.std(ddof=1) / np.sqrt(len(x)),
    mu_mle + 1.96 * x.std(ddof=1) / np.sqrt(len(x))
)
print("MLE mean:", mu_mle)
print("MLE std :", sigma_mle)
print("95% CI for mean:", ci_mu)

# 4) 부트스트랩
B = 1000
boots = [np.mean(np.random.choice(x, len(x), True)) for _ in range(B)]

# 부트스트랩 신뢰구간 (percentile method)
ci_boot = (np.percentile(boots, 2.5), np.percentile(boots, 97.5))
print("Bootstrap 95% CI:", ci_boot)
```
