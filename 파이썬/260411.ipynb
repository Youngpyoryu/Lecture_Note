# 1. 리스트 원소 개수 직접 세기

리스트를 받아 원소 개수를 반환하는 `my_len`을 만드세요.

단, len은 쓰지마세요.

> print(my_len([10,20,30,40])) #4

> print(my_len([])) #0

```python
def my_len(lst):
    count=0
    for _ in lst:
        count +=1
    return count
```

```python
my_len([10,20,30,40])
```

#2. 리스트 합계 직접 구하기

숫자 리스트를 받아 총합을 반환하는 함수 `my_sum`을 만드세요.

단, sum()은 쓰지 마세요

> print(my_sum[1,2,3,4])

> print(my_sum[15, -2,7])

```python
def my_sum(lst):
    total = 0
    for x in lst:
        total += x
    return total
```

```python
my_sum([1,2,3,4])
```

# 3. 최댓값 직접 찾기

숫자 리스트를 받아 최댓값을 반환하는 함수 `my_max`를 만드세요.

`max`는 쓰지 마세요.

> print(my_max([3, 8, 2, 10, 5]))

```python
def my_max(lst):
    largest = lst[0]
    for x in lst:
        if x>largest:
            largest = x
    return largest
```

```python
my_max([3,8,2,10,5])
```

# 4. 특정 문자의 개수 세기

문자열과 문자 하나를 받아, 그 문자가 몇 번 나오는지 세는 함수 `count_char()`를 만드세요


단, count는 쓰지 마세요.

> print(count_char("banana","a")) #3

> print(count_char("banana","b")) #1

```python
def count_char(s,target):
    count = 0
    for ch in s:
        if ch == target:
            count +=1
    return count
```

```python
count_char("banana","a")
```

```python
# find: string.find(substring, start, end)

def count_char1(s,target):
    count = 0
    start = 0

    while True:
        idx = s.find(target,start)
        if idx == -1:
            break
        count +=1
        start = idx+1
    return count

```

```python
count_char1("banana","a")
```

# 5. 리스트 뒤집기 직접 구현

리스트를 받아 뒤집힌 새 리스트 반환하는 함수 `reverse_list()`를 만드세요.

단, reversed(), 슬라이싱 [::-1]은 쓰지 마세요.

> print(reverse_list([1,2,3,4]))

```python
def reverse_list(lst):
    result = []

    last_index =0
    for _ in lst:
        last_index +=1
    last_index-=1

    while last_index >=0:
        result.append(lst[last_index])
        last_index -=1
    return result
```

```python
reverse_list([1,2,3,4,5])
```
