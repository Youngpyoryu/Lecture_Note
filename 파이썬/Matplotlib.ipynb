## Pyplot 소개

- [matplotlib.pyplot](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot) 모듈은 MATLAB과 비슷하게 명령어 스타일로 동작하는 함수의 모음.

- matplotlib.pyplot 모듈의 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있습니다.

### 기본 그래프 그리기


```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
plt.show()
```

- pyplot.plot() 함수에 하나의 숫자 리스트를 입력함으로써 아래와 같은 그래프가 그려집니다.

- plot() 함수는 리스트의 값들이 y 값들이라고 가정하고, x 값 [0, 1, 2, 3]을 자동으로 만들어냅니다.

- matplotlib.pyplot 모듈의 show() 함수는 그래프를 화면에 나타나도록 합니다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

- 스타일 지정하기.
    - <img src='https://wikidocs.net/images/page/92071/basics_07.png'>

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

- x, y 값 인자에 대해 선의 색상과 형태를 지정하는 포맷 문자열 (Format string)을 세번째 인자에 입력할 수 있습니다.

- 포맷 문자열 ‘ro’는 빨간색 (‘red’)의 원형 (‘o’) 마커를 의미합니다.

- 또한, 예를 들어 ‘b-‘는 파란색 (‘blue’)의 실선 (‘-‘)을 의미합니다.

- matplotlib.pyplot 모듈의 axis() 함수를 이용해서 축의 범위 [xmin, xmax, ymin, ymax]를 지정했습니다.

### 축 범위 지정하기

- xlim() - X축이 표시되는 범위를 지정하거나 반환합니다.
- ylim() - Y축이 표시되는 범위를 지정하거나 반환합니다.
- axis() - X, Y축이 표시되는 범위를 지정하거나 반환합니다.


<img src='https://wikidocs.net/images/page/92082/axis_range_04.png'>

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5])      # X축의 범위: [xmin, xmax]
plt.ylim([0, 20])     # Y축의 범위: [ymin, ymax]

plt.show()
```

<img src='https://wikidocs.net/images/page/92082/axis_range_05.png'>

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0, 5, 0, 20])  # X, Y축의 범위: [xmin, xmax, ymin, ymax]

plt.show()
```

### Matplotlib 그래프 영역 채우기

- fill_between() - 두 수평 방향의 곡선 사이를 채웁니다.
- fill_betweenx() - 두 수직 방향의 곡선 사이를 채웁니다.
- fill() - 다각형 영역을 채웁니다.

<img src='https://wikidocs.net/images/page/92086/fill_region_05.png'>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)

plt.show()
```

기본 사용 - fill_betweenx()

<img src='https://wikidocs.net/images/page/92086/fill_region_06.png'>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.arange(0, 2, 0.2)

# 그래프 그리기
plt.plot(x, x, 'bo', label='y = x')  # 파란색 원
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2, label='y = x^2')  # 빨간 선, 별 마커
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9, label='y = x^3')  # 초록색 선, 삼각형 마커

# x축 틱 위치와 레이블 설정 (11개의 위치와 레이블)
x_ticks = np.linspace(0, 2, 11)  # 0에서 2 사이에 11개 위치 생성
plt.xticks(x_ticks, labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])

# y축 틱 레이블 설정
plt.yticks(np.arange(0, 7), ['0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'])

# 그래프 제목과 범례 추가
plt.title("Customized Plot Example")
plt.legend()

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.arange(0, 2, 0.2)

# 그래프 그리기
plt.plot(x, x, 'bo')  # y = x (파란 원 마커)
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)  # y = x^2 (빨간 선, 별 마커)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)  # y = x^3 (초록 삼각형 마커)

# x축과 y축 레이블 설정 (레이블 개수를 10개로 조정)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', ''])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

# x축 틱 스타일 조정
plt.tick_params(axis='x',
                direction='in',
                length=3,
                pad=6,
                labelsize=14,
                labelcolor='green',
                top=True)

# y축 틱 스타일 조정
plt.tick_params(axis='y',
                direction='inout',
                length=10,
                pad=15,
                labelsize=12,
                width=2,
                color='r')

# 그래프 출력
plt.show()
```

### Matplotlib 여러 개의 그래프 그리기

<img src='https://wikidocs.net/images/page/141537/subplot_03.png'>

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 데이터 생성
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # 평균 50, 표준편차 10, 샘플 크기 100

# 평균과 표준편차 계산
mean = np.mean(data)
std = np.std(data, ddof=1)  # 샘플 표준편차

# 신뢰구간 계산 (95% 신뢰수준)
confidence_level = 0.95
z = norm.ppf((1 + confidence_level) / 2)  # Z값 계산
n = len(data)
margin_of_error = z * (std / np.sqrt(n))  # 신뢰구간의 반폭(Margin of Error)
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# 결과 출력
print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# 에러바 플롯
x = [1]
y = [mean]
yerr = [margin_of_error]  # 에러바는 신뢰구간의 반폭

plt.errorbar(x, y, yerr=yerr, fmt='o', capsize=5, label="95% Confidence Interval")
plt.title("Error Bar Representing Confidence Interval")
plt.xticks([])
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
np.random.seed(42)  # 재현성을 위해 시드 고정
data = [np.random.normal(50, 10, 200),  # 평균 50, 표준편차 10인 200개의 데이터
        np.random.normal(60, 15, 200),  # 평균 60, 표준편차 15인 200개의 데이터
        np.random.normal(70, 20, 200)]  # 평균 70, 표준편차 20인 200개의 데이터

# Boxplot 그리기
plt.boxplot(data,
            notch=True,  # 중위수에 대한 노치 표시
            patch_artist=True,  # 상자 색상 채우기 활성화
            boxprops=dict(facecolor='lightblue', color='blue'),  # 상자 속성 설정
            medianprops=dict(color='red', linewidth=2),  # 중앙값 속성 설정
            whiskerprops=dict(color='blue', linewidth=1.5),  # 수염(whisker) 속성
            capprops=dict(color='blue', linewidth=1.5))  # 끝부분 캡 속성

# x축과 y축 설정
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])  # x축 레이블
plt.ylabel('Values')
plt.title('Boxplot Example')

# 그래프 표시
plt.show()
```
