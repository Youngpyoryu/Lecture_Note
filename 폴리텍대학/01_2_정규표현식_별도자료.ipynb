## 2.1. 정규 표현식 패턴 정의하기

1. \[ \] : 괄호 안 모든 알파벳
    1. [^] : ~을 제외하고
    1. [a-z] : a~z까지 모든 알파벳  
1. *, +, \{숫자\} : 반복되는 문자열
1. ? : 있거나 없거나
1. . : 모든 문자열
    1. \d, \D : digit 숫자
    1. \w, \W : word 문자
    1. \s, \S : space 공백
1. \( \) : 패턴에서 원하는 범위만 선택

### 기본 패턴 매칭

findall을 사용하면 문자열에서 내가 찾고 싶은 단어를 검색할 수 있다.

findall(찾고 싶은 패턴, 전체 문장) : 패턴을 문장에서 모두 찾아줘

```python
pattern = r'hello'

re.findall(pattern, 'hello world!')
```

```python
pattern = '123'
string = "Phone: +1 (123) 456-7890"

re.findall(pattern, string)
```

```python
pattern = '-'
string = "Phone: +1 (123) 456-7890"

re.findall(pattern, string)
```

```python
pattern = '문자열'
string = "정규 표현식은 문자열 검색, 유효성 검사, 데이터 추출, 문자열 대체 등 다양한 작업에 유용하게 활용됩니다."

re.findall(pattern, string)
```

### 원하는 알파벳을 찾아줘 [ ]

가장 간단한 매칭 방식으로 [ ]안에 찾고자 하는 알파벳을 넣어서 매칭

ex) [abc] --> "a, b, c 중 한 개의 문자와 매치"

```python
pattern = '[abc]'

re.findall(pattern, 'before')
```

```python
pattern = '[abcde]'

re.findall(pattern, 'before')
```

```python
pattern = '[abc]'

re.findall(pattern, 'dude')
```

```python
pattern = '[123]'
string = "Phone: +1 (123) 456-7890"

re.findall(pattern, string)
```

```python
string = 'apple app aple apppple a00e aqqle ale APPLE Apple (apple) [apple]'
```

```python
re.findall('.', string)  #.은 모든 글자(character) '하나'를 나타냅니다.
```

```python
re.findall('a.p', string)  #a와 l 사이에 아무 글자 하나가 있는 모든 경우

# re.findall('a.l', string)
#aple가 아니라 apl이 나온걸 주의
```

```python
re.findall('a[.]p', string)
```

#### 이것은 제외하고!

^를 쓰면 뒤에오는 것을 제외하라는 의미입니다. (not)

```python
re.findall('a[^q]le', string)  #a와 le 사이에 q를 제외한 단어가 있을 때
```

#### - 이것부터 ~ 저기까지

-를 쓰면 시작과 끝 사이 모든 것을 찾을 수 있습니다.

- 0-9 : 모든 숫자
- a-z : 모든 소문자
- a-zA-Z : 모든 알파벳
- 가-힣 : 모든 한글

```python
string = "Phone: +1 (123) 456-7890"

re.findall('[1-5]', string)  #1~5까지 숫자를 찾아줘
```

```python
example = 'class는 attribute와 method로 이루어져 있습니다.'
print(re.findall('[a-zA-Z]', example))
print(re.findall('[가-힣]', example))
```

### 몇 번이나 나올까?

앞에서 [  ]만으로는 너무 낱개로 잘라져서 나오죠?..

정규표현식에는 해당 패턴이 몇 번 나오는지 정할 수 있는 몇 가지 방법이 있습니다.

#### ? - 있느냐 없느냐

해당 글자가 있든 없든 매칭시킵니다.

```python
re.findall('ap?le', string)
```

#### {} - 몇 번 반복되느냐

{n} 또는 {n, m}은 몇 번 반복되는지 지정할 수 있습니다.

```python
re.findall('ap{2}le', string)  #p가 두번 반복되는 경우만
```

```python
re.findall('ap{2,5}le', string)  #p가 2~5번 반복되는 경우만
```

#### + - 1번 이상 반복되느냐?

+를 쓰면 1번 이상 반복되는 패턴을 찾을 수 있습니다.

```python
re.findall('ap+le', string)
```

#### * - 0번 이상 반복되느냐?

*는 +와 유사하지만 없는 경우, 즉 0번 이상 반복패턴을 찾을 수 있습니다.

```python
re.findall('ap*le', string)
```

### 문자 클래스

문자 클래스는 공백, 숫자, 글자등을 대표하는 표기법입니다.

- \d (digit) 숫자
- \D 숫자가 아닌거
- \s (space) 공백
- \S 공백이 아닌거
- \w (word) 글자
- \W 글자가 아닌거

### 같이 사용하기



```python
example = 'class는 attribute와 method로 이루어져 있습니다.'
print(re.findall('[가-힣]+', example))
print(re.findall('[a-zA-Z]+', example))
print(re.findall('\w+', example))
```

```python
# phone = "Phone: +1 (123) 4546-7890"
phone = "Phone: +82 (055) 4563-7890"
# phone = "Phone: +82 (02) 3449-7890"

print(re.findall('[0-9]+', phone))  #1~5까지 숫자를 찾아줘
print(re.findall('\+\d+', phone))  #+가 앞에 붙은 숫자를 찾아줘
```

### 그루핑 - 내가 원하는 일부만

패턴으로 찾았지만 뽑고 싶은건 그 중 일부만일 수 있습니다.

(  )를 사용하면 패턴 중에서 원한느 범위만 추출할 수 있습니다.

```python
aca = "솔데스크의 SOL은 Solomon왕의 지혜처럼 우리의 인생을 성공하기 위해(Success of Life) 다양한 해결방안(Solution)을 제공하는 회사가 되길 바라는 의미입니다."

print(re.findall('\((.+?)\)', aca))  #괄호 속 text를 구하고 싶다.
print(re.findall('\w*c{2}\w*', aca))  #c가 두 번 이상 반복된 단어
```

```python
phone = "Phone: +82 (055) 4563-7890"
print(re.findall('\((\d+)\)', phone))  #+가 앞에 붙은 숫자를 찾아줘
print(re.findall('(\d+)-(\d+)', phone))  #-양쪽에 있는 숫자들을 찾아줘
```
